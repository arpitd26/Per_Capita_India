```python
import pandas as pd 
import geopandas as gpd
import matplotlib.pyplot as plt 

df = pd.read_excel('Capita.xlsx')

df
```

```python
shp_gdf = gpd.read_file('India_State_Boundary.shp')
shp_gdf
```

```python
merged = shp_gdf.set_index('State_Name').join(df.set_index('States/Union Teritory'))
merged
```

```python
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Per Capita Income of India 2022',
             fontdict={'fontsize': '15', 'fontweight' : '3'})
fig = merged.plot(column='2021-22', cmap='RdYlGn', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```
